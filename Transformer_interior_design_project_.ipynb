```python
import torch
import gradio as gr
from PIL import Image
from diffusers import StableDiffusionPipeline

# Check if GPU is available
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load Stable Diffusion Text-to-Image Model
text_to_img_pipe = StableDiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2-1", torch_dtype=torch.float16
).to(device)

# Function to generate a realistic interior design image from a prompt
def generate_realistic_interior(prompt):
    realistic_image = text_to_img_pipe(prompt).images[0]
    realistic_image = realistic_image.convert("RGB")  # Ensure correct format
    return realistic_image

# Gradio UI
with gr.Blocks() as demo:
    gr.Markdown("# üè° AI-Powered Interior Design Generator")

    with gr.Row():
        text_input = gr.Textbox(label="Enter Interior Prompt", placeholder="Example: A modern minimalist living room")
        generate_btn = gr.Button("Generate Realistic Interior")

    realistic_output = gr.Image(label="Generated Realistic Interior")

    generate_btn.click(
        fn=generate_realistic_interior,
        inputs=text_input,
        outputs=realistic_output
    )

demo.launch(debug=True)
```
